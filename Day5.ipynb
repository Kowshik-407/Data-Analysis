# Combining

```python
dict1 = {'Name':['Kowshik407','Aravind415','Sridhar405'],
        'Branch':['CSE','ECE','CSSE']}
df1 = pd.DataFrame(dict1)
df1
```

```python
dict2 = {'Name':['Jassu434','Triyekh436','Teja447'],
        'Branch':['MECH','CIVIL','IT']}
df2 = pd.DataFrame(dict2)
df2
```

```python
pd.concat([df1,df2])
```

- axis = 0  --> It represent rows
- axis = 1  --> It represent cols

```python
pd.concat([df1,df2], axis = 1)
```
