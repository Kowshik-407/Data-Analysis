## jointplot

```python
# Combination of two plots is known as jointplot
sns.jointplot(x="sepal_length",y='sepal_width',data=iris,kind='reg')
plt.show()
```

```python
sns.jointplot(x="petal_length",y='petal_width',data=iris,kind='reg')
plt.show()
```

```python
sns.jointplot(x="sepal_length",y='sepal_width',data=iris)
plt.show()
```

```python
sns.jointplot(x="petal_length",y='petal_width',data=iris)
plt.show()
```

```python
sns.jointplot(x="sepal_length",y='sepal_length',data=iris,kind='kde')
plt.show()
```

```python
sns.jointplot(x="petal_length",y='petal_length',data=iris,kind='kde')
plt.show()
```

```python
sns.jointplot(x="petal_length",y='sepal_length',data=iris,kind='kde')
plt.show()
```

```python
sns.jointplot(x="sepal_length",y='petal_length',data=iris,kind='kde')
plt.show()
```

#### Try for widths of both sepal and petal

```python
sns.jointplot(x="petal_length",y='sepal_length',data=iris,color='k').plot_joint(sns.kdeplot,n_levels=7)
plt.show()
```

```python
help(sns.jointplot)
```

```python
x , y = np.random.randn(2,300)
sns.jointplot(x,y,kind='hex')
plt.show()
```

```python
sns.jointplot(x="sepal_length",y='sepal_width',data=iris,kind='hex')
plt.show()
```

```python
sns.jointplot(x="sepal_length",y='sepal_width',data=iris,kind='scatter')
plt.show()
```

```python
sns.jointplot(x="sepal_length",y='sepal_width',data=iris,kind='resid')
plt.show()
```

## pairplot

```python
# It is used to plot multiple plots at a time for the entire plot.

sns.set_style("ticks")
sns.pairplot(iris,hue="species",diag_kind="kde",kind="scatter",palette="husl")
plt.show()
```

```python
sns.pairplot(iris,hue="species",markers=['o','d','s'])
# Let us save this fig with the name seaborn.png
plt.savefig("Seaborn.png")
plt.show()
```

## stripplot

```python
sns.stripplot(x="species",y='sepal_length',data=iris)
plt.show()
```

## heatmap

```python
a = np.random.randn(10,20)
g = sns.heatmap(a)
plt.show()
```

```python
sns.heatmap(a,vmin=0,vmax=-5)
plt.show()
```

```python
plt.figure(figsize=(10,8)) # It will increase the figure
sns.heatmap(a,annot=True)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(a,annot=True,linewidth=0.5)
plt.show()
```

```python
sns.heatmap(a,cmap='YlGnBu')
plt.show()
```
