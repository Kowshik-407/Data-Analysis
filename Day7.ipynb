## Subplots

- subplot is a function used to creae multiple plots at a time.
- subplot(number_of_rows, number_of_columns)

### Creating Subplots

```python
plt.subplots(2,2)
plt.show()
```

```python
plt.subplot(2,2,2)
plt.show()
```

```python
fig,a = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('Square Graph')
a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title('SquareRoot Graph')
a[1][0].plot(x, np.exp(x))
a[1][0].set_title('Exponential Graph')
a[1][1].plot(x, np.log2(x))
a[1][1].set_title('Logarithmic Graph')

# plt.tight_layout()
# plt.show()
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan=2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)

x = np.arange(1,10)

a1.plot(x,x)
a1.set_title('Linear Plot')
a1.set_xlabel('X-axis')
a1.set_ylabel('Y-axis')
a2.plot(x,x**2)
a2.set_title('Square Plot')
a2.set_xlabel('X-axis')
a2.set_ylabel('Y-axis')
a3.plot(x,x**3)
a3.set_title('Cubic Plot')
a3.set_xlabel('X-axis')
a3.set_ylabel('Y-axis')

a1.grid()
a2.grid(True)
a3.grid(color='m',ls='-.')

plt.tight_layout()
plt.show()
```
